```python
# split train dataset into train and validation set
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

x_train, x_val, y_train, y_val = train_test_split(x_train, y_train, test_size=0.2, random_state=0)
```

analyze random forest model's feature importances to select features

```python
# import libraries
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from lightgbm import LGBMClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score,plot_confusion_matrix,confusion_matrix
import matplotlib.pyplot as plt
```

```python
# default random forest 
clf = RandomForestClassifier(random_state=42)
clf.fit(x_train, y_train)
```

```python
# function for analyzing features of trained-random-forest model
def feature_importance():
  plt.figure(figsize=(6, 4))
  feat_importances = pd.Series(clf.feature_importances_, index=peminer_header)
  feat_importances.nlargest(30).plot(kind='bar')

  # nonzero_feat_importances: features with feature_importances larger than zero
  nonzero_feat_importances = {}
  for i in range(0,188):
    if feat_importances.values[i] != 0:
      nonzero_feat_importances[feat_importances.index[i]] = feat_importances.values[i]

  plt.ylabel("score")
  return nonzero_feat_importances

nonzero_feat_importances = feature_importance()
```

```python
# colList: features with zero feature_importances 
# this is for dropping columns
colList = []
for i in peminer_header:
  if i not in nonzero_feat_importances:
    colList.append(i)
print(colList)
print(len(colList))
```

```python
x_train_drop = x_train.drop(colList, axis=1)  # drop columns
x_val_drop = x_val.drop(colList, axis=1)  # drop columns
x_test_drop = x_test.drop(colList, axis=1)  # drop columns
```

define utility function

```python
# training algorithms
def load_model(**kwargs):
  if kwargs["model"] == "rf":
    return RandomForestClassifier(random_state=42, n_jobs=4)
  elif kwargs["model"] == "dt":
    return DecisionTreeClassifier(random_state=42)
  elif kwargs["model"] == "lgb":
    return LGBMClassifier(random_state=42)
  elif kwargs["model"] == "svm":
    return SVC(random_state=42)
  elif kwargs["model"] == "rfc_nonzero":
    return rfc_nonzero
  else:
    return None
```

```python
# function for training, default: random forest 
def train(x_train, y_train, model="rf"):
  clf = load_model(model=model, random_state=42)  # load model passed by parameter
  clf.fit(x_train, y_train)  # train
  return clf

# function for evaluating and visualizing
def evaluate(clf, x_test, y_test):
  predict = clf.predict(x_test)
  # visualize confusion matrix
  accuracy = clf.score(x_test, y_test)  # evaluate accuracy
  print(f"Accuracy: {accuracy}")
  f_measure = f1_score(y_test, predict, average=None, labels=[0, 1])
  print(f"F_measure: {f_measure}")
  recall = recall_score(y_test, predict, average=None, labels=[0, 1])
  print(f"Recall: {recall}")
  precision = precision_score(y_test, predict, average=None, labels=[0, 1])
  print(f"Precision: {precision}\n\n")
  plot_confusion_matrix(clf, x_test, y_test)
  plt.show()
```

## 2. Random Forest, Decision Tree, LGBMClassifier, SVM itself

```python
# train
models = []
for model in ["rf", "dt", "lgb", "svm"]:
    clf = train(x_train, y_train, model)
    models.append(clf)

for model in models:
  print(model)
  evaluate(model, x_val, y_val)  # visualization
```

## 3. Ensemble

```python
models = []
for model in ["rf", "lgb"]:  # for ensemble, we chose rf and lgb (by accuracy)
    clf = train(x_train, y_train, model)
    models.append(clf)

for model in models:
  print(model)
  evaluate(model, x_val, y_val)  # visualization
```

```python
# ensemble trained model 
# soft voting
def ensemble_result(X, y, models):
    predicts = []
    for model in models:
        prob = [result for _, result in model.predict_proba(X)]
        predicts.append(prob)
    
    predict = np.mean(predicts, axis=0)
    predict = [1 if x >= 0.5 else 0 for x in predict]  
    print("accuracy", accuracy_score(y, predict))
```

```python
ensemble_result(x_val, y_val, models)
```

```python
# test data ensemble
def ensemble_data(x_test, models):
    predicts = []
    for model in models:
        prob = [result for _, result in model.predict_proba(x_test)]
        predicts.append(prob)
    
    predict = np.mean(predicts, axis=0)
    predict = [ 1 if x >= 0.5 else 0 for x in predict ]
        
    return predict
  
result = ensemble_data(x_test, models)
```

## 5. RNN

```python
# RNN
from keras.models import Sequential
from keras.layers import Dense, LSTM, Embedding, Dropout, SimpleRNN
from keras.layers import SpatialDropout1D

rnn = Sequential()
# First LSTM layer
rnn.add(LSTM(units = 10,activation='tanh', return_sequences = True, input_shape = (x_train.shape[1], 1)))
# RNN layers and Dropout regularizations
rnn.add(SimpleRNN(units = 50,activation='tanh', return_sequences = True))
rnn.add(Dropout(0.2))
rnn.add(SimpleRNN(units = 50,activation='tanh', return_sequences = True))
rnn.add(Dropout(0.2))
rnn.add(SimpleRNN(units = 50,activation='tanh', return_sequences = True))
rnn.add(Dropout(0.2))
rnn.add(SimpleRNN(units = 50))
rnn.add(Dropout(0.2))
rnn.add(Dense(units = 1)) # output layer

# compile RNN
rnn.compile(optimizer = 'adam', loss = 'mean_squared_error' ,metrics=['accuracy'])

# fit RNN to the training set
t = rnn.fit(x_train, y_train, epochs = 30, batch_size = 256, validation_data=(x_val, y_val))
```

```python
# Visualize the result of RNN
plt.plot(t.history['accuracy'], label='train')
plt.legend(loc='upper left')
plt.title('Model accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.show()
plt.plot(t.history['loss'], label='train')
plt.legend(loc='upper right')
plt.title('Model Cost')
plt.ylabel('Cost')
plt.xlabel('Epoch')
plt.show()
```
